### Definición de clases <a name="clases"></a>

Ya que los datos obtenidos no están agrupados, se diseñó una clase que tiene como objetivo regresar una tabla ordenada en la que se contengan todas las estadísticas por temporada, las cuales son:
- **Team** : Equipo participante en la temporada especificada
- **Points**: Cantidad de puntos final del equipo en la temporada especificada
- **Win Games**: Cantidad de juegos ganados del equipo en la temporada especificada
- **Draw Games**: Cantidad de juegos empatados del equipo en la temporada especificada
- **Lost Games**: Cantidad de juegos perdidos del equipo en la temporada especificada
- **Favor Goals**: Cantidad de goles a favor del equipo en la temporada especificada
- **Against Goals**: Cantidad de goles recibidos del equipo en la temporada especificada
- **Diff**: Balance de goles del equipo en la temporada especificada
- **Season**: Temporada especificada

Esta clase devuelve un objeto del tipo DataFrame en el que se contiene la información antes mencionada.

```python
# TODO hacer clase para formar las tablas de las diferentes temporadas
class SeasonStats:
    # Constuctor clase SeasonStats
    def __init__(self):
        self.teams = None
        self.table = None
        self.season = ""
    
    # Metodo set usado para especificar la temporada
    def set_season(self, season):
        self.season = season
    
    # Metodo get usado para obtener la temporada que usamos
    def get_season(self):
        return self.season
    
    # Metodo usado para obtener la tabla ordenada de la temporada con la informacion de cada equipo 
    def get_team_stats_by_season(self, data, team):
        team = np.array(team)
        favor_goals = []
        against_goals = []
        diff = []
        wins = []
        loss = []
        draws = []
        points = []
        for i in range(len(team)):
            mask = ((data['home_team'] == team[i]) | (data['away_team'] == team[i])) & (data['season'] == self.get_season())
            data_temp = data[mask]
            home_games = data_temp[data_temp['home_team'] == team[i]]
            away_games = data_temp[data_temp['away_team'] == team[i]]
            wins.append((home_games.result == 'H').sum() + (away_games.result == 'A').sum())
            loss.append((home_games.result == 'A').sum() + (away_games.result == 'H').sum())
            draws.append((home_games.result == 'D').sum() + (away_games.result == 'D').sum())
            favor_goals.append(int(home_games['home_goals'].sum() + away_games['away_goals'].sum()))
            against_goals.append(int(data_temp['home_goals'].sum() + data_temp['away_goals'].sum()) - favor_goals[i])
            diff.append(int(favor_goals[i] - against_goals[i]))
            points.append(wins[i]*3 + draws[i]*1)
            
        self.table = { 
            'Team': team,
            'Points': points,
            'Win Games': wins,
            'Draw Games': draws,
            'Lost Games': loss,
            'Favor Goals': favor_goals,
            'Against Goals': against_goals,
            'Diff': diff,
            'Season': np.array([self.get_season()]*len(team))
        }
        self.table = pd.DataFrame(self.table)
        self.table.sort_values(by= ['Points', 'Diff'], inplace= True, ascending= False, ignore_index= True)
        # self.table = self.table.style.apply(self.highlight_row, axis=1)
        return self.table
    
    # Metodo usado para resaltar en la tabla al campeon y los equpipos a descender
    def highlight_row(self, data):
        if data.name == 0:
            return ['background-color: yellow']*8
        
        elif (data.name >= len(self.teams)-3 and data.name < len(self.teams)):
            return ['background-color: red']*8
    
    # Metodo usado para traer todos los equipos que participan en la temporada especificada
    def get_all_teams_by_season(self, data):
        season = self.get_season()
        
        if data[data['season'] == season].empty:
            return None
        
        data = data[data['season'] == season]
        self.teams = data
        return self.teams
```

- Liverpool en la temporada 2017-2018 fue la 2da mejor ofensiva y la 4ta mejor defensiva.
- Southampton en la temporada 2014-2015 aunque fue la 2da mejor defensiva, no estuvo ni cerca de estar entre el top 3 de mejores ofensivas.

En ambos casos podemos observar como estos equipos inclinaron su fortaleza hacia un lado u otro de la cancha, siendo en Liverpool la ofensiva y en Southampton la defensiva. Al analizar esto, podemos llegar a la siguiente conclusión: Es importante tener un balance tanto en la ofensiva como en la defensiva.

De nuevo, tomemos como ejemplo la temporada 2017-2018 para demostrar lo anterior. En la temporada antes mencionada, Manchester City fue el equipo campeón con la mejor ofensiva y la mejor defensiva, además de obtener el récord de más puntos en una sola temporada con 100 puntos.

Como se mencionó anteriormente, podemos ver que un equipo que guarda una relación positiva entre su defensiva y ofensiva tiene mayores posibilidades de estar entre los mejores equipos en la tabla al final de la temporada, y mientras mejor sea esta relación, mejores serán las posibilidades. Esto demuestra que no se puede depender solamente de un delantero, defensa o portero para tener éxito en la temporada, y es un error muy común entre muchos equipos que invierten cantidades millonarias en delanteros que garantizan una cifra importante de goles a favor pero descuidan reforzar la defensiva o portería.

### Predicción de Puntos en relación a diferencial de goles <a name="Predicción"></a>

### Análisis de la Distribución de Probabilidad del Diferencial de Goles  <a name="Distribución"></a>

Antes de tratar de entrenar algún modelo, es importante entender los datos que utilizaremos para nuestro experimento. Primero haremos un histograma para observar si nuestra variable Diff sigue alguna distribución. 

```python
plt.hist(all_time_stats.Diff, bins=11, color='b')
plt.title("# de equipos por diferencia de goles")
plt.xlabel("Diferencia de goles")
plt.ylabel("# de equipos")
plt.show()
```

Al observar el histograma resultante, podemos observar que nuestra variable Diff sigue aproximadamente una Distribución Normal. La Distribución Normal es un modelo que nos permitirá aproximar el valor de una variable aleatoria a una situación real, y a través de una función de probabilidad. Para nuestro ejemplo, la función de probabilidad nos ayudará a predecir qué porcentaje de equipos se encuentran en cierto rango de diferencia de goleo.

Del anterior gráfico podemos decir lo siguiente:
- La mayoría de equipos se encuentran cerca de la media de un diferencial de 0 goles
- Los equipos con peor diferencial de goles se encuentran a la izquierda de la medida central o media
- Los equipos con mejor diferencial de goles se encuentran a la derecha de la media central o media

Podemos comprobar que el diferencial de cero goles es la media de la siguiente manera:

```python
mean_diff = all_time_stats.Diff.mean() 
std_diff = all_time_stats.Diff.std() 

print(f'Media en el diferencial de goles de los equipos: {mean_diff}') 
print(f'Desviación estandar en el diferencial de goles de los equipos: {std_diff}')
```

Ya que la distribución normal es una distribución que tiene forma de campana, es importante evaluar la precisión de nuestro histograma y que tan cerca está de ser una distribución normal. Recordando la teoría de la Distribución Normal, tenemos la siguiente regla:

$$ mean(x) \pm 1 \sigma \approx 68.3% $$
$$ mean(x) \pm 2 \sigma \approx 95.5% $$
$$ mean(x) \pm 3 \sigma \approx 99.7% $$

Lo que nos dice esta regla es que aproximadamente el 68.3% de los valores están dentro de la primera sigma, 95.5% dentro de la segunda sigma y 99.7 dentro de la tercera. Ahora, para comprobar que nuestra variable aleatoria **Diferencial de Goles** sigue una distribución normal, haremos el cálculo de las tres sigmas:

```python
# Cálculo de la las 3 sigmas 
def get_sigmas(data, variable, sigma, standard_dev, mean):
    return (data[(data[variable] >= (mean - sigma*standard_dev)) & (data[variable] <= mean + sigma*standard_dev)][variable]).count()

sigma_1 = get_sigmas(all_time_stats, 'Diff', 1, std_diff, mean_diff)
sigma_2 = get_sigmas(all_time_stats, 'Diff', 2, std_diff, mean_diff)
sigma_3 = get_sigmas(all_time_stats, 'Diff', 3, std_diff, mean_diff)


diff_total_count = all_time_stats.Diff.count()
# approx 68%
percentage_std1 = sigma_1 / diff_total_count * 100
# approx 95%
percentage_std2 = sigma_2 / diff_total_count * 100
# approx 99.7%
percentage_std3 = sigma_3 / diff_total_count * 100 

print(f'Sigma 1% : {percentage_std1}')
print(f'Sigma 2% : {percentage_std2}')
print(f'Sigma 3% : {percentage_std3}')
```

Al observar el resultado obtenido, podemos concluir que nuestra variable aleatoria **Diferencial de Goles** tiene una distribución normal, ya que el cálculo de las 3 sigmas es bastante aproximado a los valores esperados. Es importante hacer la aclaración de que es una aproximación, por lo que si dibujamos la función de probabilidad, no será 100% normal y tendremos casos que no se ajusten a esta función, sin embargo, nos servirá para el estudio que haremos de esta variable.

En el siguiente histograma podemos ver la función de probabilidad correspondiente a nuestra variable aleatoria:

```python
# Histograma 
plt.hist(all_time_stats.Diff, bins=11, density=True, alpha=0.6, color='g')
plt.title("# de equipos por diferencia de goles")
plt.xlabel("Diferencia de goles")
plt.ylabel("# de equipos")
 
# Graficar la función de probabilidad
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean_diff, std_diff)
plt.plot(x, p, 'k', linewidth=3, color='r')
title = "# de equipos por diferencia de goles. Promedio = %.2f,  std = %.2f" % (mean_diff, std_diff)
plt.title(title)
plt.show()
```

### Introducción Regresión Lineal <a name="regresion_introduccion"></a>

Una vez que ya hemos modelado la función de probabilidad de la variable aleatoria **Diferencial de goles** podemos proceder a la predicción de valores. Cómo ya hemos mencionado anteriormente, el objetivo de esta predicción será el de obtener la cantidad posible de puntos en base al diferencial de goles por equipo. Ya que este es un problema donde tendremos como salida un valor numérico, usaremos algoritmos de aprendizaje supervisado para obtener la predicción. 

Los algoritmos de aprendizaje supervisado nos serán de gran ayuda, ya que este es un problema donde tenemos etiquetas y queremos saber la cantidad de puntos por equipo en un periodo de tiempo. Existen dos tipos de problemas que estos algoritmos pueden resolver:

1. Problemas de clasificación
2. Problemas de regresión

Ya que deseamos obtener una variable numérica como salida, utilizaremos algoritmos de regresión, específicamente, haremos una regresión lineal para obtener la relación entre nuestra variable aleatoria diferencial de goles y los puntos obtenidos por temporada.

De manera resumida, una regresión lineal es una técnica de modelado estadístico que nos permite crear un modelo lineal. Este modelo describe la relación entre una variable dependiente $y$ como una función de una o varias variables independientes $Xi$. Para implementar la regresión lineal, usaremos la librería sklearn de Python.

Para entrenar nuestro modelo de regresión lineal haremos lo siguiente: 
1. Dividiremos nuestro dataset en dos partes comprendidas por el 70% y el 30%.
2. Entrenaremos nuestro modelo con el dataset que contiene el 70% del dataset original 
3. A partir del modelo entrenado vamos a predecir los puntos obtenidos con el dataset que contiene el 30% del dataset original
4. Mediremos la eficiencia de nuestro algoritmo

### Implementación Regresión Lineal <a name="regresion_implementacion"></a>

```python
# Dividimos el dataset en las variables train y test para entrenar el modelo de regresión lineal
train, test = train_test_split(all_time_stats, test_size=0.30, random_state=50)

# Asignamos las variables que vamos a estudiar en la regresión
xtrain = train[['Diff']]
ytrain = train[['Points']]

xtest = test[['Diff']]
ytest = test[['Points']]

# Iniciar un objeto del tipo LinearRegression
regr = linear_model.LinearRegression()
# Entrenar el modelo 
regr.fit(xtrain, ytrain)

# Hacemos predicción de puntos obtenidos usando el dataset test
y_pred = regr.predict(xtest)

# Graficamos la regresión obtenida  
plt.scatter(xtest, ytest,  color='black')
plt.plot(xtest, y_pred, color='blue', linewidth=3)
plt.xlabel("Diferencial de Goles")
plt.ylabel("Puntos obtenidos")
plt.show()
```

###  Eficiencia del modelo <a name="eficiencia_modelo"></a>

Cómo ya se ha mencionado anteriormente, existen una variedad de algoritmos para resolver este tipo de problemas, y una parte importante del análisis y modelado es medir la eficiencia para escoger el método más preciso. Para medir la eficiencia de nuestro modelo, utilizaremos dos métodos:

1. Error cuadrado medio
2. Coeficiente de determinación

El error cuadrado medio va a medir el cuadrado del error que tenemos entre nuestra predicción y el valor esperado y mientras más pequeño sea el valor obtenido, mejor será nuestro modelo. Por otro lado, el coeficiente de determinación de igual manera medirá la eficiencia de nuestro modelo y obtendremos valores entre 0 y 1. Mientras más cerca esté de 1, mejor será nuestro modelo.

```python
# Error cuadrático medio
print("Error cuadrado medio: %.2f" % mean_squared_error(ytest, y_pred))

# Coeficiente de determinación
print('Coeficiente de determinación: %.2f' % r2_score(ytest, y_pred))
```

Al medir la eficiencia de nuestro modelo, obtenemos un valor de 19.80 para el error cuadrático medio y 0.94 para el coeficiente de determinación. Podemos decir que nuestro modelo tiene una muy buena calidad, ya que aunque habrá un error en los valores predecidos, este error no afectará mucho el resultado.

Finalmente, para probar la calidad en los valores obtenidos de nuestro modelo, agregaremos al dataset test las predicciones y el Error entre las predicciones y el valor esperado. Como se mencionó anteriormente, aunque existe un ligero error en los valores obtenidos, este error no desvía mucho el resultado.

```python
# Agregamos al dataset test las columnas 'Points Prediction'y 'Error'
pd.options.mode.chained_assignment = None
test['Points Prediction'] = y_pred
test['Error'] = np.abs((test.Points - test['Points Prediction']) / test.Points * 100)
display(test[['Team', 'Points', 'Points Prediction', 'Error', 'Diff']].head())
```

Para terminar con nuestro experimento de modelación, probaremos la regresión lineal con nuestro dataset original. Esto nos ayudará a obtener predicciones de un mayor dataset y medir la calidad del modelo.

```python
# Predecir los puntos obtenidos utilizando el dataset completo 
y_all = regr.predict(all_time_stats[['Diff']])

# Graficamos la regresión obtenida  
plt.scatter(all_time_stats['Diff'], all_time_stats['Points'],  color='black')
plt.plot(all_time_stats['Diff'], y_all, color='blue', linewidth=3)
plt.xlabel("Diferencial de Goles")
plt.ylabel("Puntos obtenidos")
plt.show()

# Error cuadrático medio
print("Error cuadrado medio: %.2f" % mean_squared_error(all_time_stats[['Points']], y_all))

# Coeficiente de determinación
print('Coeficiente de determinación: %.2f' % r2_score(all_time_stats[['Points']], y_all))
```

Podemos observar que tanto el Error cuadrático medio como el coeficiente de determinación nos dio muy buenos valores, y aunque existe un pequeño error, el modelo entrenado nos da valores bastante aproximados a los esperados.
