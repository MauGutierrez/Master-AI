#### **Sobre el conjunto de datos:**


El conjunto de datos que analizaremos es un conjunto de datos para el reconocimiento de escritura a mano: cada observación corresponde a una imagen de 8 por 8 bits donde cada variable representa la intensidad pixel. En total dispondremos de 64 variables de entrada. A su vez cada imagen tendrá una etiqueta categórica (0 a 9). Sin embargo, el análisis de componentes principales no requiere de las etiquetas ya que no vamos a clasificar cada imagen nueva. PCA proyectará las entradas del espacio de 64 dimensiones en un espacio más pequeño, reduciendo así el coste del problema y abriendo más enfoques de aprendizaje automático.

#### **Repaso teórico del PCA:**

El análisis de componentes principales es un mapeo del espacio original de alta dimensión de  *$d$*  variables de entrada a un subespacio de dimensiones inferiores de  *$k$* variables. La intención es minimizar la pérdida de información (maximizar la cantidad de variación en los datos explicada por los componentes principales, mientras se minimiza el número de componentes principales).


pca.png


Es importante señalar que PCA es una técnica de reducción de dimensionalidad; no usa las salidas. El PCA se utiliza para reducir la dimensionalidad y, por lo tanto, se combina con otras técnicas de aprendizaje automático que pueden utilizar el conjunto de variables de entrada de dimensiones inferiores.

En términos de álgebra lineal, estamos proyectando un vector de entrada  **$X$** a un vector de entrada reducido  **$Z$**  usando una matriz de proyección  **$W$** :

**$$Z =W^TX$$**
 
La construcción de la matriz de proyección se basa en un problema de maximización: maximizar la varianza explicada por cada componente principal. El componente principal es **$W_1$** , un vector de longitud *$d$* , y es una combinación ponderada de variables de entrada que, después de que los datos se proyectan en **$W_1$**, distribuye los datos al máximo.

El problema de optimización se convierte entonces en: para una observación dada  *$i$* , maximiza la varianza del vector  **$Z_i$**, que es igual a:

$Var(Z_i) =W^T_iΣW_i$
 
mientras restringe la norma L2 de cada vector de proyección a 1,

**$W^T_iW_i= 1$**
 
Esto se reduce a unas pocas operaciones de álgebra lineal, que haremos a mano, y luego con el objeto PCA de scikit-learn.

#### **Normalización de los datos:**

Nuestros datos de entrada son un poco desordenados: cuadrículas de 8x8 de valores enteros con una escala no estándar. Es útil escalar las variables de entrada usando un *StandardScaler*. Esto escalará las variables para que tengan una media cero y una desviación estándar de aproximadamente 1.

Aquí hay un método que toma una matriz de variables de entrada como parámetro y devuelve la matriz de entradas estandarizada, un valor medio para las entradas y una matriz de covarianza:

```python
def get_normed_mean_cov(X):
    X_std = StandardScaler().fit_transform(X)
    X_mean = np.mean(X_std, axis=0)
    
    ## Automatic:
    #X_cov = np.cov(X_std.T)
    
    # Manual:
    X_cov = (X_std - X_mean).T.dot((X_std - X_mean)) / (X_std.shape[0]-1)
    
    return X_std, X_mean, X_cov

X_std, X_mean, X_cov = get_normed_mean_cov(X_training)
print(X_std.shape)
print(X_mean.shape)
print(X_cov.shape)
```

#### **Tratamiento de los datos:**

Ahora que tenemos la matriz de covarianza, podemos visualizarla. Esto nos informa sobre la covarianza entre diferentes píxeles en las imágenes de 8x8. Vemos visualmente la estructura de las imágenes de 8x8: las imágenes que están en los extremos opuestos de la imagen tienen covarianzas negativas, lo que significa que es probable que estén en lados opuestos de la escala superior al promedio / inferior al promedio. Recuerde la definición de covarianza:

**$Cov(ψ,ϕ)=∑_{i=1}^N\frac{(ψi−\bar{ψ})(ϕi−\bar{ϕ})}{N}$**

```python
fig = plt.figure(figsize=(12,12))
sns.heatmap(pd.DataFrame(X_cov), annot=False, cmap='PuOr')
plt.show()
```

#### **Autovectores y autovalores de la matriz de covarianza:**


```python
eigenvals, eigenvecs = np.linalg.eig(X_cov)

# Los valores propios no se ordenan necesariamente, pero el valor_propio[i] corresponde a vector_propio[i]

print("Dimensiones valor propio: "+str(eigenvals.shape))
print("Dimensiones vector propio: "+str(eigenvecs.shape))
```

```python
index = np.argsort(eigenvals)
index = index[::-1]
# ordenar los autovectores acorde con los autovalores ordenados:
eigenvecs = eigenvecs[index]
# Ordenando los autovalores de mayor a menor:
eigenvals = eigenvals[index]
```

Podemos visualizar los autovectores, ordenados por rango de autovalor, e identificar visualmente qué componentes de autovector dominan en cada autovalor. Esto es esencialmente una visualización de qué información el análisis de componentes principales ha considerado más importante.



```python
# Create a tuple of (eigenvalues, eigenvectors)
unsrt_eigenvalvec = [(np.abs(eigenvals[i]), eigenvecs[:,i]) for i in range(len(eigenvals))]

# Sort tuple by eigenvalues
eigenvalvec = sorted(unsrt_eigenvalvec, reverse=True, key=lambda x:x[0])

#pprint([pair for pair in eigenvalvec])
```

```python
fig = plt.figure(figsize=(14,3))
sns.heatmap(pd.DataFrame([pair[1] for pair in eigenvalvec[0:21]]), 
            annot=False, cmap='coolwarm',
           vmin=-0.5,vmax=0.5)

plt.ylabel("Valores propios por orden")
plt.xlabel("Autovectores")
plt.show()
```

Más de la estructura cíclica de las imágenes presentes en el mapa de calor anterior. Los puntos de valores altos y bajos indican que puede haber algunos valores donde la presencia de un píxel es un fuerte indicador de que un número es o no un dígito determinado. #### **Explicación de la varianza:**

Para medir el "principado" de cada componente principal - qué tan importante es - podemos determinar qué parte de la varianza en los datos que explica. (Tenga en cuenta que aquí decimos varianza porque representar los datos usando la media se considera la línea de base con la que se compara todo lo demás. (Esta es la misma idea detrás del coeficiente R cuadrado de Pearson). Si hay más varianza (variación de la media) en los datos puede ser explicado por el modelo, el modelo es mejor.

La varianza explicada individual del componente principal  $k$  se denota  $λ_k$ .

La varianza explicada del  $k^{th}$  componente principal (dado  $D$  componentes principales totales) viene dado por:

$\frac{λ_1 + λ_2 + ⋯ + λ_k}{λ_1+λ_2+ ⋯ +λ_k+ ⋯ +λ_D}$
 
Comenzaremos calculando las variaciones explicadas individuales,  $λ_k$ .


```python
lam_sum = sum(eigenvals)
explained_variance = [(lam_k/lam_sum) for lam_k in sorted(eigenvals, reverse=True)]
```

La distribución de la varianza explicada para cada componente principal da una idea de cuánta información se representará y cuánto se perderá cuando se reduzca la entrada completa de 64 dimensiones utilizando un modelo de componente principal (es decir, un modelo que utiliza solo el primero $N$  componentes principales).

```python
plt.figure(figsize=(6, 4))

plt.bar(range(len(explained_variance)), explained_variance, alpha=0.5, align='center',
        label='Contribución individual de cada $\lambda_{k}$')

plt.ylabel('Ratio de la varianza explicada')
plt.xlabel('Componentes principales ordenados')

plt.legend(loc='best')
plt.tight_layout()
```

```python
fig = plt.figure(figsize=(6,4))
ax1 = fig.add_subplot(111)

ax1.plot(np.cumsum(explained_variance))

ax1.set_ylim([0,1.0])

ax1.set_xlabel('Número de componentes principales')
ax1.set_ylabel('Acumulación de la varianza explicada')
ax1.set_title('Varianza explicada')

plt.show()
```

Nuestra intención con el análisis de componentes principales es reducir la entrada de alta dimensión a una entrada de baja dimensión. En última instancia, esa entrada de baja dimensión está pensada para su uso en un modelo, que entrenaremos con los datos de entrenamiento y probaremos con los datos de prueba.

#### **Proyección al subespacio $K$ con scikit-learn**


Para transformar los puntos de 64 entradas de alta dimensión en el subespacio de dimensiones inferiores de los componentes principales (el nuevo subespacio de características), ponderamos cada uno de los valores de las variables de entrada de acuerdo con el vector propio. Esto se hace usando una matriz de proyección, que es una matriz que se multiplica por el vector de entrada para cambiar su tamaño en un subespacio más pequeño.

Por ejemplo, para transformar un conjunto de vectores de entrada de 64 dimensiones  $X$  en sus versiones subespaciales aproximadas de 4 dimensiones  Z  usando la matriz de proyección  $W$ , necesitaremos hacer la siguiente multiplicación de matrices:

$Z = W^TX$
 
dónde  $W^T$  es un  4 × 64  matriz de proyección, y  $X$  es un  64 × $N$ matriz de las entradas originales (esta es la transposición de lo que hemos almacenado en la $X_{std}$ variable).

Esto resultará en una matriz  $Z$  eso es de tamaño  4 × $N$. Luego, esta matriz se puede usar como un conjunto reducido de variables de entrada y se pueden aplicar una variedad de otras técnicas de aprendizaje automático. El costo de estas técnicas a menudo aumenta a medida que O ($N^2$), por lo que la reducción de 64 dimensiones a 4 puede conducir a disminuciones de órdenes de magnitud en el coste computacional de los modelos de aprendizaje automático.

#### **Evaluación de un modelo de PCA: error de reconstrucción y retroproyección**


Hemos proyectado las entradas originales de alta dimensión en un subespacio de menor dimensión, utilizando la matriz de proyección:

$Z = W^TX$
 
Es posible evaluar qué tan bueno es el modelo usando retroproyección, para transformar las entradas de baja dimensión nuevamente en un espacio de alta dimensión, con cierta pérdida de información. El conjunto retroproyectado de entradas de alta dimensión se denota $\hat{X}$ :

$\hat{X}= WZ$
 
Tenga en cuenta que esto utiliza el hecho de que  $WW^T= I$ , como se implementó en las restricciones de optimización. La distancia euclidiana entre el original y los puntos retroproyectados proporciona una medida de cuánta información se perdió:

$∥\hat{X}- X∥^2$
 
De hecho, al utilizar PCA, estamos seleccionando componentes principales de modo que cada componente principal adicional minimice esta cantidad.

Al comparar visualmente las entradas originales y retroproyectadas, podemos examinar visualmente qué información se pierde y cuánta, utilizando imshow()para mostrar los datos originales y retroproyectados.

```python
matW = np.hstack(pair[1].reshape(64,1) for pair in eigenvalvec[0:4])
X_digit1 = X_std[0]
Z_digit1 = X_digit1.dot(matW)
Xhat_digit1 = Z_digit1.dot(matW.T)

fig = plt.figure(figsize=(10,6))
ax1, ax2= (fig.add_subplot(221+i) for i in range(2))



ax1.imshow(X_digit1.reshape(8,8))
ax1.set_title('Imagen original')

ax2.imshow(Xhat_digit1.reshape(8,8))
ax2.set_title('Imagen retropropagada')

plt.show()
```
